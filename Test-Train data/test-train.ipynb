```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("test-train.xlsx")
print(df)
```

```python
plt.scatter(df['milage'], df['sellprice'], marker='o')
plt.title("Milage vs sellprice")
plt.xlabel("milage")
plt.ylabel("sellprice")
```

```python
plt.scatter(df['age '], df['sellprice'], marker='o')
plt.title("age vs sellprice")
plt.xlabel("age")
plt.ylabel("sellprice")
```

```python
x= df[['milage','age ']]
y= df['sellprice']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
reg.score(x_test, y_test)  #score is used to check the accuracy of the model
```
