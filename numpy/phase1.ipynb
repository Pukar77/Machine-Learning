array and basics


```python
import numpy as np
```

Creating array from list

```python
array_1d = np.array([1,2,3,4,5])
print(array_1d)
print(array_1d[2])
```

```python
array_2d = np.array([[1,2,3], [4,5,6]])
print(array_2d)
```

Numpy VS List Array

```python

py_list = [1,2,3,4,5]
print("List Multipliation", py_list * 2)

np_array = np.array([1,2,3,4,5])
print("Numpy multiplication", np_array * 2)

#Numpy operation is faster than list operation
```

Creating matrix

```python
ones = np.ones((2,3))
print(ones, "\n")

zeros = np.zeros((4,5))
print("\n",zeros)

full = np.full((2,2),9)
print("\n",full)

random = np.random.random((2,3))
print("\n", random)

sequence = np.arange(0, 10, 3)
print(sequence)


```

Vector, Matrix and Tensor
