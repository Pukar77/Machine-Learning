```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("houseprice.xlsx")
df
```

```python
plt.plot(df["area"], df["price"], marker='o')  #for bargraph you can use, plt.bar
plt.xlabel("Area")
plt.ylabel("price")
plt.title("Area vs house")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df['price'])  #fit method ma input rw targeted output dinxa, jasma input must be 2D array, that is why we used [[]], and for output it is same as others
```

```python
reg.predict([[3300]])
```
